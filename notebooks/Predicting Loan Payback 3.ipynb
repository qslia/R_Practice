* **Loads** the newly installed package
* `suppressMessages()` hides startup messages
* `character.only = TRUE` treats `pkg` as a string

## `character.only = TRUE` - Why Is This Needed?

Without `character.only = TRUE`:

```r
# This WORKS (direct package name):
library(dplyr)

# This FAILS (variable containing name):
pkg <- "dplyr"
library(pkg)  # Error! R looks for a package called "pkg"

# This WORKS (with character.only):
pkg <- "dplyr"
library(pkg, character.only = TRUE)  # Correct!
```

**Explanation:**
- `library(dplyr)` looks for a package literally named "dplyr"
- `library(pkg)` looks for a package literally named "pkg" (not the value of `pkg`)
- `library(pkg, character.only = TRUE)` uses the **value** stored in `pkg` variable

## What This Code Does

This line performs a **two-step conversion** to safely transform the `loan_paid_back` column into numeric format.

---

## Why Two Steps? The Factor Problem

This double conversion is necessary when your column is a **factor** (a special R data type for categorical data). Let me show you why:

### The Problem: Direct Conversion from Factor

If `loan_paid_back` is a factor containing values like "0" and "1", here's what happens:

```r
# Suppose loan_paid_back is a factor
loan_paid_back <- factor(c("0", "1", "1", "0", "1"))
print(loan_paid_back)
# [1] 0 1 1 0 1
# Levels: 0 1

# WRONG WAY - Direct conversion
as.numeric(loan_paid_back)
# [1] 1 2 2 1 2  â† WRONG! Converted to level numbers, not actual values!
```

**Problem:** `as.numeric()` on a factor returns the **internal level codes** (1, 2), not the actual values (0, 1)!

---

## The Solution: Two-Step Conversion

### Step 1: `as.character(train$loan_paid_back)`

First, convert the factor to **character strings**:

```r
# Factor â†’ Character
as.character(loan_paid_back)
# [1] "0" "1" "1" "0" "1"  â† Now they're text strings
```

This gives you the **actual labels** ("0", "1"), not the internal codes.

---

### Step 2: `as.numeric(...)`

Then convert the character strings to **numbers**:

```r
# Character â†’ Numeric
as.numeric(as.character(loan_paid_back))
# [1] 0 1 1 0 1  â† Correct! Actual numeric values
```

Now you have true numeric values (0, 1) that can be used in mathematical operations and modeling.

---

## Visual Comparison

| Original Data Type | Direct Conversion | Two-Step Conversion |
|-------------------|-------------------|---------------------|
| Factor: "0", "1", "1", "0" | `as.numeric()` â†’ 1, 2, 2, 1 âŒ | `as.numeric(as.character())` â†’ 0, 1, 1, 0 âœ… |

## Why Is This Needed for Loan Prediction?

In binary classification problems (like loan payback prediction):

- **0** typically means: Loan **NOT** paid back (default)
- **1** typically means: Loan **paid back** successfully

Machine learning models like XGBoost require:
- âœ… Numeric format (0, 1)
- âŒ Not factors
- âŒ Not character strings

## What This Code Does

This function converts **categorical variables** into **binary (0/1) dummy variables** suitable for machine learning models like XGBoost.

---

## Function Breakdown

### `fastDummies::dummy_cols()`

* **Package:** `fastDummies`
* **Function:** `dummy_cols()`
* **Purpose:** Create dummy/indicator variables from categorical columns
* **Syntax:** `package::function()` explicitly calls the function from that package

---

## Parameters Explained

### 1. **`train`** - The Data

The original dataframe containing both numeric and categorical columns.

---

### 2. **`select_columns = cat_cols`** - Which Columns to Encode

* **`cat_cols`** is a vector of categorical column names (defined earlier in your code)
* Only these columns will be converted to dummy variables
* Other columns remain unchanged

**Example:**
```r
cat_cols <- c("employment_status", "loan_purpose", "education_level")
# Only these 3 columns will be dummy-encoded
```

---

### 3. **`remove_first_dummy = TRUE`** - Avoid Multicollinearity

This is **crucial** for regression models!

#### Without `remove_first_dummy` (default FALSE):

Suppose you have a column `employment_status` with 3 categories:

| Original | Dummy 1 | Dummy 2 | Dummy 3 |
|----------|---------|---------|---------|
| Employed | 1 | 0 | 0 |
| Self-employed | 0 | 1 | 0 |
| Unemployed | 0 | 0 | 1 |

**Problem:** These 3 columns are **perfectly collinear**:
- If `Dummy1=0` and `Dummy2=0`, then `Dummy3` MUST be 1
- `Dummy1 + Dummy2 + Dummy3` always equals 1
- This causes the "**dummy variable trap**" in regression models

#### With `remove_first_dummy = TRUE`:

| Original | Dummy 2 | Dummy 3 |
|----------|---------|---------|
| Employed | 0 | 0 |
| Self-employed | 1 | 0 |
| Unemployed | 0 | 1 |

**Solution:** Drop the first dummy!
- First category becomes the **reference/baseline** (all dummies = 0)
- No multicollinearity
- Models can estimate coefficients properly

**Formula:**
- **n categories** â†’ **n-1 dummy variables** (when `remove_first_dummy = TRUE`)
- **n categories** â†’ **n dummy variables** (when `remove_first_dummy = FALSE`)

---

### 4. **`remove_selected_columns = TRUE`** - Clean Up

* Removes the **original categorical columns** after creating dummies
* Keeps your dataframe clean with only encoded variables

#### Comparison:

**With `remove_selected_columns = FALSE`:**
```
| employment_status | employment_status_Self-employed | employment_status_Unemployed | ... |
|-------------------|--------------------------------|------------------------------|-----|
| Employed          | 0                              | 0                            | ... |
| Self-employed     | 1                              | 0                            | ... |
```
â† Original column still there (redundant!)

**With `remove_selected_columns = TRUE`:**
```
| employment_status_Self-employed | employment_status_Unemployed | ... |
|---------------------------------|------------------------------|-----|
| 0                               | 0                            | ... |
| 1                               | 0                            | ... |
```
â† Original column removed (cleaner!)

## Common Settings Comparison

| Parameter | TRUE | FALSE |

| `remove_first_dummy = TRUE` | n-1 dummies (avoid collinearity) âœ… | n dummies (redundant) |

### For Linear/Logistic Regression:
